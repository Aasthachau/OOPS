##Q.4: What is the difference between abstraction and encapsulation

ANSWER: Core Difference: What vs. How--


**Abstraction** focuses on what the object does. (The Design Level)

**Encapsulation** focuses on how the object protects its data. (The Implementation Level)

**Abstraction:-** Hiding Complexity
Abstraction is a design concept focused on simplifying the interface presented to the user.

Goal: To manage complexity by showing only essential features and hiding all unnecessary implementation details.

Focus: What the object is capable of doing.

Implementation: Achieved through techniques like Abstract Classes and Interfaces, which define a contract for behavior without providing the complete code.

Real-World Analogy (Driving a Car): The steering wheel, accelerator, and brake pedal are an abstraction. They let you interact with the car's movement without needing to know the complex internal mechanics of the engine or transmission.


**Encapsulation:-** Hiding Data
Encapsulation is an implementation mechanism focused on bundling and protection.

Goal: To protect the object's data (attributes) by restricting direct access and ensuring data integrity.

Focus: How the data and code are wrapped into a single unit (the class).

Implementation: Achieved by using access modifiers (like private) to hide variables and exposing them only through controlled methods (Getters and Setters).

Real-World Analogy (Driving a Car): The engine and transmission are encapsulated under the hood. Their complex inner workings are hidden and protected from external, unauthorized tinkering, ensuring the car functions reliably

##Q.6: Explain the concept of inheritance in OOP

ANSWER: Inheritance is one of the four foundational pillars of Object-Oriented Programming (OOP). It is a mechanism that allows a new class to inherit (or adopt) properties and behaviors from an existing class.

1. The Core Concept
Inheritance establishes a hierarchical, parent-child relationship between classes:

Parent Class (Base Class or Superclass): The existing class whose features are being inherited.

Child Class (Derived Class or Subclass): The new class that inherits features from the parent class.

The key conceptual relationship established by inheritance is the "is-a" relationship.

Example: A Dog is a type of Animal. A Sedan is a type of Car.

2. Benefits and Purpose
The primary reasons for using inheritance are:

Code Reusability: The child class automatically gets all the non-private attributes and methods of the parent class, meaning you don't have to write the same code twice.

Polymorphism: Inheritance is essential for implementing polymorphism, allowing a single method name to be used across different classes in the hierarchy.

Logical Organization: It helps organize classes into a clear, intuitive hierarchy that models the real-world domain.

3. How Inheritance Works
Consider the relationship between an Animal class and a Dog class:

Animal Class (Superclass): Defines common attributes (age, color) and methods (eat(), sleep()).

Dog Class (Subclass):

It inherits the age, color, eat(), and sleep() features automatically.

It adds new, specific features that only dogs have, such as the breed attribute and the bark() method.

This means you only define the eat() method once in the Animal class, and every class that inherits from Animal (like Dog, Cat, Bird) can use it without any extra code.

4. Method Overriding
A child class can also override a method inherited from the parent class. This means the child class provides its own specific implementation for a method that is already defined in the parent.

Example: Both Dog and Cat inherit the make_sound() method from Animal. However, the Dog class can override it to print "Woof," and the Cat class can override it to print "Meow." The method name is the same, but the behavior is specific to the object type.

##Q.8: How is encapsulation achieved in Python

ANSWER--> Encapsulation: in the context of Python, is achieved through a combination of naming conventions and the use of properties (Getters and Setters), rather than strict, compiler-enforced access modifiers like those found in languages like Java or C++.

1. The Role of Naming Conventions (Non-Public Members):

Python uses special underscore prefixes to indicate that a method or attribute is intended for internal use and should not be accessed directly from outside the class.

A. Single Leading Underscore (_attribute)-->


Convention: Prepending an attribute or method name with a single underscore (_) signals that the member is "protected" or non-public.


Purpose: It is a programmer-to-programmer warning. It tells other developers: "This is part of the internal implementation of the class; access or modify it at your own risk."

B. Double Leading Underscore (__attribute)-->


Mechanism: Prepending with a double underscore (__) triggers a mechanism called name mangling.


Access: This makes direct external access more difficult and cumbersome, effectively making the member "private" by confusing its name. However, it can still be accessed indirectly if the mangled name is known.


2. Controlling Access with Properties (@property):
The most robust and Pythonic way to implement controlled encapsulation is by using properties.

This involves creating:

A "Private" Attribute: Using the single underscore convention (e.g., self._temperature).

A Getter Method: Defined using the @property decorator to control how the attribute is read.

A Setter Method: Defined using the @attribute.setter decorator to control how the attribute is changed. This is where you can add validation logic.

Purpose:

To group a utility function logically within the class, where the function has some conceptual relationship to the class but does not need to access the class state (class variables) or the instance state (instance variables).

##Q.11: What is method overloading in Python

In Object-Oriented Programming (OOP), method overloading is a form of compile-time polymorphism that allows a class to have multiple methods with the same name but different signatures (i.e., different numbers or types of parameters).

The Python Reality
The most important point is that Python does not support method overloading in the traditional sense.

Here is the theory behind why Python behaves this way:

Dynamic Typing: Python is dynamically typed. It does not check function signatures during the compile phase based on the number or type of arguments.

Overwrite Mechanism: When the Python interpreter encounters multiple methods with the exact same name within a single class definition, the subsequent definition simply overwrites the previous one.

Pythonic Alternatives (Achieving the Effect)
Although you cannot define two separate methods with the same name, Python's flexible argument handling allows you to achieve the effect of overloading using a single method definition:

Default Arguments: By giving some parameters a default value (like None), you make those arguments optional. The single method can then handle different numbers of arguments (e.g., calling it with two arguments or three arguments).

Variable-Length Arguments (*args): Using the *args syntax allows the method to accept an arbitrary (any) number of positional arguments. The method can then check the number of items received and adapt its logic.

Type Checking: Inside the single method, you can use conditional logic (if/elif/else) and functions like isinstance() to check the type of the received arguments. The method's behavior can then change based on the data type it receives.


##Q.12: What is method overriding in OOP

ANSWER: Method Overriding is a key feature of Runtime Polymorphism in Object-Oriented Programming (OOP). It allows a subclass (child class) to provide a specific implementation for a method that is already defined in its superclass (parent class).


1. The Core Principle
Method overriding works exclusively within an Inheritance hierarchy.

(A)Rule: The method signature (name, number, and type of parameters) in the subclass must exactly match the method signature in the superclass.

(B)Goal: To enable objects of different classes, which share a common ancestor, to respond to the same method call with behavior unique to their own type.

2. How it Works
(A)Inheritance: A Subclass inherits a method (e.g., move()) from its Superclass.

(B)Definition: The Superclass provides a general implementation for move() (e.g., "Generic movement").

(C)Overriding: The Subclass then redefines the same move() method.

(D)Runtime Decision: When the method is called on an object of the subclass, the subclass's version of the method is executed, effectively replacing the superclass's version for that specific object.

3. Why Use Overriding?


Specific Behavior: Allows subclasses to provide behavior specific to their type while still adhering to the general contract (the method signature) defined by the parent.

Plug-and-Play: Enables lists or arrays of mixed object types (e.g., a list of Animal objects) to be processed using a loop, where the same method call on each item yields the correct, type-specific result.

##Q.13: What is a property decorator in Python

ANSWER--> The @property decorator is a built-in feature in Python that serves a critical role in achieving encapsulation—one of the core pillars of Object-Oriented Programming (OOP).

1. The Core Purpose: Attribute Control:  
The primary function of the @property decorator is to transform a method inside a class into an attribute that can be accessed without parentheses. This is often called unifying the interface or separating the interface from the implementation.

2. The Mechanism of Controlled Access:  
The property mechanism is comprised of methods that share the same base name, controlling different access types:

The Getter (@property): This is the main definition. It is the method that executes when a user reads the value of the property (e.g., my_object.attribute). This method is often used to calculate a value or return a formatted version of the internal data.

The Setter (@name.setter): This is the method that executes when a user attempts to assign a new value to the property (e.g., my_object.attribute = value)

The Deleter (@name.deleter): This optional method executes when the user attempts to delete the property using the del keyword.

3. Key Benefits:  


Encapsulation: It protects the internal state of the object by forcing any attempt to change the data to go through the setter method, where validation and safety checks are performed.

Clean Interface: It allows users to interact with the object using simple attribute access (object.attribute), even though complex logic might be running behind the scenes.

Flexibility: You can refactor the internal workings of the class (how the data is stored) without changing the external code that uses the class.

##Q.17: What is the difference between a class variable and an instance variable

Class Variable:

1. Belongs to the class, not to a specific object.

4. Declared inside the class, outside methods.

5. Used to store common data for all objects.

1. Belongs to a specific object (instance).

2. Each object has its own separate copy.

3. Multiple copies exist (one per object).

4. Declared inside methods (usually the constructor).

5. Used to store object-specific data.

6. Change in value affects only that object.

##Q.2:

Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both.

```python
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Child class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Child class Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Create objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())
```

```python
# Parent class
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

    def show_type(self):
        print(f"Vehicle type: {self.type}")

# Child class
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def show_brand(self):
        print(f"Car brand: {self.brand}")

# Grandchild class
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery):
        super().__init__(vehicle_type, brand)
        self.battery = battery

    def show_battery(self):
        print(f"Battery capacity: {self.battery} kWh")


my_car = ElectricCar("Car", "Tesla", 100)


my_car.show_type()     # Vehicle method
my_car.show_brand()    # Car method
my_car.show_battery()  # ElectricCar method
```

##Q.5:

 Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Amount deposited:", amount)
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print("Amount withdrawn:", amount)
        else:
            print("Insufficient balance or invalid amount")

    def check_balance(self):
        print("Current balance:", self.__balance)


# Object creation
account = BankAccount(1000)

account.deposit(500)
account.withdraw(300)
account.check_balance()
```

method subtract_numbers() to subtract two numbers

```python
class MathOperations:

    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b


# Method calls
sum_result = MathOperations.add_numbers(10, 5)
sub_result = MathOperations.subtract_numbers(10, 5)

print("Addition:", sum_result)
print("Subtraction:", sub_result)
```

##Q.10:

Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"


# Object creation
v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2   # operator overloading
print("Resultant Vector:", v3)
```

##Q.14:

```python
class Employee:
    def calculate_salary(self, hours_worked, hourly_rate):
        return hours_worked * hourly_rate


class Manager(Employee):
    def calculate_salary(self, hours_worked, hourly_rate, bonus):
        basic_salary = super().calculate_salary(hours_worked, hourly_rate)
        return basic_salary + bonus


# Object creation
emp = Employee()
mgr = Manager()

print("Employee Salary:", emp.calculate_salary(40, 500))
print("Manager Salary:", mgr.calculate_salary(40, 500, 10000))
```

##Q.16:

Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass


class Cow(Animal):
    def sound(self):
        print("Cow says Moo")


class Sheep(Animal):
    def sound(self):
        print("Sheep says Baa")


# Object creation
c = Cow()
s = Sheep()

c.sound()
s.sound()
```
